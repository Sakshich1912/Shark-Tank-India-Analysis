```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import statistics as stat
import warnings
warnings.filterwarnings('ignore')
```

```python
df_sharktank=pd.read_excel('EDA_Shark_Tank_India.xlsx')
```

```python
#Q1. Find the number of episodes?

num_episodes = df_sharktank['episode_number'].nunique()
print(f"Number of episodes: {num_episodes}")
```

```python
#Q2. Max, min , mean of asked amount,asked equity, asked valuation? 

asked_amount_stats = df_sharktank['pitcher_ask_amount'].agg(['max', 'min', 'mean'])
asked_equity_stats = df_sharktank['ask_equity'].agg(['max', 'min', 'mean'])
asked_valuation_stats = df_sharktank['ask_valuation'].agg(['max', 'min', 'mean'])

print("Asked Amount Stats:\n", asked_amount_stats)

print("Asked Equity Stats:\n", asked_equity_stats)

print("Asked Valuation Stats:\n", asked_valuation_stats)
```

```python
#Q3. Max & Min asked- equity,asked-valuation and asked amount episode-wise? 

episode_stats = df_sharktank.groupby('episode_number').agg({
    'pitcher_ask_amount': ['max', 'min'],
    'ask_equity': ['max', 'min'],
    'ask_valuation': ['max', 'min']
})
episode_stats
```

```python
#Q4. Brand names in which 2 ,3 or 4 sharks are invested? 

brands_with_sharks = df_sharktank[df_sharktank['total_sharks_invested'].isin([2, 3, 4])]['brand_name'].unique()

brands_with_sharks
```

```python
#Q5. Episode wise minimum and maximum of sharks invested? 

episode_sharks_stats = df_sharktank.groupby('episode_number')['total_sharks_invested'].agg(['min', 'max'])

episode_sharks_stats
```

```python
#Q7. Brand names who ask for 1 crore and got a deal? 

df_sharktank[(df_sharktank['pitcher_ask_amount'] == 100) & (df_sharktank['deal'] == 1)]['brand_name'].unique()

```

```python
#Q8. Brand names where deal equity is between 0 to 50? 

df_sharktank[(df_sharktank['deal_equity'] >= 0) & (df_sharktank['deal_equity'] <= 50)]['brand_name'].unique()

```

```python
#Q9. Find the number of brands participated in each episode? 

df_sharktank.groupby('episode_number')['brand_name'].nunique()
```

```python
#Q10. How many sharks participated in this show and What were their names? 

shark_columns = ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']
num_sharks = len(shark_columns)

# Extract and capitalize shark names
shark_names = [shark.split('_')[0].capitalize() for shark in shark_columns]

print(f"Number of sharks: {num_sharks}")
print("Names of sharks:", ", ".join(shark_names))
```

```python
#Q11. Find appearance of each sharks? 
# Define shark columns
shark_columns = ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']

shark_appearance_df = df_sharktank[shark_columns].apply(pd.Series.value_counts).fillna(0)
shark_appearances = shark_appearance_df.loc[1]
shark_appearances
```

```python
#Q12. How many entrepreneurs were present? 
num_entrepreneurs = df_sharktank['brand_name'].nunique()
print(f"Number of entrepreneurs: {num_entrepreneurs}")
```

```python
#Q13. How many times each shark invested the deal? 
shark_investments_count = df_sharktank[df_sharktank['deal'] == 1][shark_columns].apply(pd.Series.value_counts).loc[1]
shark_investments_count = shark_investments_count.astype(int).to_dict()
print("Each shark investment count:", shark_investments_count)
```

```python
#Q14. Find the equity percent that each sharks gets? 
# Filter the DataFrame for deals that were made (deal == 1)
deals_made = df_sharktank[df_sharktank['deal'] == 1]

# Initialize a dictionary to store equity shares for each shark
shark_total_equity = {}

# Iterate over shark columns
for shark in shark_columns:
    # Calculate total equity for each shark
    shark_name = shark.split('_')[0].capitalize()  # Extract shark name from column
    total_equity = deals_made[deals_made[shark] == 1]['equity_per_shark'].sum()
    shark_total_equity[shark_name] = total_equity

print("Total equity for each shark:")
print(shark_total_equity)
```
